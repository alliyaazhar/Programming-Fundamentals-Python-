# Problem # 1: (Room Tiles Calculator)

```python
def main():
    print('---------------------------------------------------')
    room=int(input('number of rooms to be covered : '))
    print('---------------------------------------------------')
    print('note : 1 foot has 12 inches.') 
    print('---------------------------------------------------')
    tile=int(input('size of each tile (in inches) : '))
    while tile not in range(1,13):
        tile=int(input('please enter tile size between 1 to 12 inches : '))
    print('---------------------------------------------------------------------------------')
    tilecal(room,tile)
    
    

def tilecal(room,tile):
    totltotal=0
    for i in range (1,room+1):
        lenn=[]
        widd=[]
        print('enter the length of room',i,'(in feet and inches separated by a coma) : ',end='')
        lenth=eval(input())
        lenn+=lenth
        l0=lenn[0]
        l1=lenn[1]
        lt=l0+(l1/12)
        print('enter the width of room',i,'(in feet and inches separated by a coma) : ',end='')
        width=eval(input())
        widd+=width
        w0=widd[0]
        w1=widd[1]
        wt=w0+(w1/12)
        area=lt*wt
        etile=area//tile
        if area%tile!=0:
            etile+=1
        print('----------------------------------------------------------------------------------')
        print('number of tiles needed for this room : ',etile)
        print('-------------------------------------------------')
        totltotal+=etile
    end(totltotal)
        
def end(totltotal):
    print('total number of tiles needed : ',totltotal)
    boxes=totltotal//20
    if totltotal%20!=0:
        boxes+=1
    print('total number of boxes neended : ',boxes)
    print('-----------------------------------------------------')
    
    
main()
```

# Problem # 3 (Exam Grading App)

```python
def enterdata():
    outfile = open('exams.txt', 'a')
    per = 'y'
    infile = open('scores.txt', 'w')
    infile.write("STUDENT'S ID \t STUDENT'S SCORE \n")
    infile.close()
    while per == 'y':
        std_id = input("enter student's ID : ")
        ans = input("enter student's solution : ")
        outfile.write(std_id)
        outfile.write('\t \t')
        outfile.write(ans)
        outfile.write('\n')
        writing_scores(std_id, ans)
        per = input('do you want to keep adding? y/n : ')
    outfile.close()


def writing_scores(std_id, ans):
    key = 'abcdefabcdefabcdefab'
    infile = open('scores.txt', 'a')
    infile.write(std_id)
    infile.write('\t \t')
    score = len(ans)
    if 'g' or 'h' or 'i' or 'j' or 'k' or 'l' or 'm' or 'n' or 'o' or 'p' or 'q' or 'r' or 's' or 't' or 'u' or 'v' or 'w' or 'x' or 'y' or 'z' in ans:
        infile.write('(Invalid Answers)')
    else:
        if score == 20:
            grade = sum(1 for i in range(score) if ans[i] == key[i])
            infile.write(str(grade))
        if score < 20:
            infile.write('(Too Few Answers)')
        if score > 20:
            infile.write('(Too Many Answers)')
    infile.write('\n')
    infile.close()


def main():
    outfile = open('exams.txt', 'w')
    outfile.write('------------------------------------------------- \n')
    outfile.write('|key == > a b c d e f a b c d e f a b c d e f a b|\n')
    outfile.write('--------------------------------------------------\n')
    outfile.write('STUDENT ID \t SOLUTIONS  \n')
    outfile.write('--------------------------------------------------\n')
    outfile.close()
    enterdata()


main()
```
